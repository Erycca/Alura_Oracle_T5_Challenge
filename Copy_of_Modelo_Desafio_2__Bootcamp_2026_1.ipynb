DESAFIO 1

Você recebeu uma lista com os títulos das palestras submetidas para um evento "Women in Data". Sua primeira tarefa é fazer uma contagem rápida de quantos títulos mencionam temas de BI.

Crie um programa que percorra a lista de títulos e conte quantos deles contêm as palavras-chave "dados", "análise" ou "inteligência" (sem diferenciar maiúsculas e minúsculas).

```python
titulos_palestras = [
    "Machine Learning para Negócios",
    "O Poder da Análise de Dados no Varejo",
    "Introdução ao SQL",
    "Business Intelligence com Power BI",
    "Carreira em Dados: Dicas para Mulheres",
    "Storytelling com Inteligência Artificial"
]
```

```python
#Execute seu código aqui

palavras_chaves = ["dados", "análise", "inteligência"]

contador = 0

for titulo in titulos_palestras:
  titulo_lower = titulo.lower()

  for palavra_chave in palavras_chaves:
    if palavra_chave in titulo_lower:
      contador += 1
      break

print("Quantidade de Palestras Sobre BI:", contador)
```

DESAFIO 2

Como parte de uma análise inicial, você precisa criar uma função para avaliar um conjunto de valores numéricos.

Crie uma função que receba uma lista de números e retorne:

o valor máximo

o valor mínimo

a média dos valores

```python
#Execute seu código aqui

numeros = [10, 5, 6, 20, 15]

def analisar_valores(numeros):

  valor_maximo = max(numeros)
  valor_minimo = min(numeros)
  media = sum(numeros) / len(numeros)

  return valor_maximo, valor_minimo, media
print(analisar_valores(numeros))

```

DESAFIO 4

Após uma palestra sobre "O Futuro da IA", foi feita uma enquete com a pergunta: "Qual foi o tópico mais impactante?". Os resultados foram armazenados em um dicionário onde a chave é o tópico e o valor é o número de votos.

Sua tarefa é calcular o número total de votos e exibir o resultado da enquete em porcentagens.

```python
resultado_enquete = {
    "IA Generativa": 120,
    "Ética em IA": 85,
    "IA no Mercado de Trabalho": 95,
    "Ferramentas de MLOps": 50
}
```

```python
#Execute seu código aqui

total_votos = sum(resultado_enquete.values())

print("Total de Votos:", total_votos)
print()

for topico, votos in resultado_enquete.items():
  porcentagem = (votos / total_votos) * 100
  print(f"{topico}: {porcentagem:.2f}% ({votos} votos)")
```
